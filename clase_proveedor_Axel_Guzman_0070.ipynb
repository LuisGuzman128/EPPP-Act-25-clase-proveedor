```python
# Act24 clase Proveedor Axel Guzman 0070
# Programa interactivo para el registro y uso de 2 proveedores

class Proveedor:
    def __init__(self, nombre, empresa, telefono):
        self.nombre = nombre
        self.empresa = empresa
        self.telefono = telefono
        self.registro_validado = False

    # Funciones de la clase
    def leerdatos(self):
        self.registro_validado = True
        print(f"📡 [SISTEMA]: Sincronizando datos de contacto con la empresa {self.empresa}...")

    def mostrardatos(self):
        if self.registro_validado:
            print(f"👤 [PROVEEDOR]: {self.nombre} | Empresa: {self.empresa} | Teléfono: {self.telefono}")
        else:
            print(f"⚠️ [ERROR]: Los datos de {self.nombre} aún no se han sincronizado en el sistema.")

# --- BLOQUE INTERACTIVO: REGISTRO DE DATOS ---

# Registro del Primer Proveedor
print("=== REGISTRO PROVEEDOR 1 ===")
n1 = input("Nombre del contacto: ")
e1 = input("Empresa proveedora: ")
t1 = input("Teléfono: ")
proveedor1 = Proveedor(n1, e1, t1)

# Registro del Segundo Proveedor
print("\n=== REGISTRO PROVEEDOR 2 ===")
n2 = input("Nombre del contacto: ")
e2 = input("Empresa proveedora: ")
t2 = input("Teléfono: ")
proveedor2 = Proveedor(n2, e2, t2)

# --- EJECUCIÓN DE RESULTADOS ---

print("\n--- PROCESANDO PROVEEDOR 1 ---")
proveedor1.leerdatos()
proveedor1.mostrardatos()

print("\n--- PROCESANDO PROVEEDOR 2 ---")
# En este no ejecutamos leerdatos() para probar el mensaje de error
proveedor2.mostrardatos()

print("\n-----------------------")
print("Programa terminado")
print("creador de software por Axel Guzman NL= 0070")
```
